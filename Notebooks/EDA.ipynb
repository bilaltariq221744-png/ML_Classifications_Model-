#### Dataset Overview

This dataset is designed for analyzing student behavior, academic performance, mental health factors, and dropout risk prediction. It contains 1,000,000 samples, making it suitable for large scale machine learning experiments.

The primary goal of this dataset is to understand how academic, psychological, and lifestyle factors contribute to student dropout risk.

### Features Description

The dataset includes the following categories of features:

## Academic & Demographic Features
- age: Age of the student
- gender: Gender category of the student
- academic_year: Current academic year or semester level
- academic_performance: Academic performance indicator
- exam_pressure: Level of academic exam pressure experienced

## Study Behavior Features
- study_hours_per_day: Average daily study hours
- screen_time: Daily screen usage time
- internet_usage: Amount of internet usage per day

## Mental Health Indicators
- stress_level: Perceived stress level
- anxiety_score: Anxiety measurement score
- depression_score: Depression severity score
- burnout_score: Burnout level indicator
- mental_health_index: Overall mental health score

## Lifestyle & Well-being Features
- sleep_hours: Average sleep duration per day
- physical_activity: Level of physical activity

## Social & Environmental Factors
- social_support: Level of social support available
- family_expectation: Pressure from family expectations
- financial_stress: Financial stress level

## Target Variables
- risk_level: Categorical risk classification (e.g., Low, Medium, High)
- dropout_risk: Continuous dropout risk score representing likelihood of student dropout

## Worse mental health -> higher chance of dropping out


A student studies more to do well but that also increases exam pressure. Over time pressure turns into stress which brings along anxiety and sometimes depression. These build up and lead to burnout.
If the student is not sleeping enough or staying active things get worse. Sports from others can help a bit while money worries and family expectations add more stress.As all this piles up their mental health drops and if burnout and stress get too high the risk of dropping out increases.

## Conclusions and Key Takeaways

### 1. **Dataset Characteristics**
- The dataset contains 1,000,000 student records with a comprehensive mix of academic, behavioral, psychological, and socioeconomic features
- **Data Quality**: No missing values or duplicates detected; well-structured dataset
- **Class Imbalance**: Severe imbalance in target variable (76.7% Low, 21.8% Medium, 1.5% High risk) - requires careful handling during model training

### 2. **Key Findings from Correlation Analysis**
- **Mental Health Cluster**: Strong interconnections between stress (0.76 with anxiety), anxiety (0.48 with depression), and burnout (0.75 with stress)
- **Strongest Dropout Predictors**: 
  - Burnout score (0.69 correlation)
  - Depression score (0.65)
  - Stress level (0.55)
  - Mental health index (-0.63, inverse relationship)
- **Protective Factors**: Sleep hours and social support show negative correlations with stress and depression

### 3. **Demographic Insights**
- Age and gender distributions are balanced, limiting predictive power individually
- Mental health and stress levels remain relatively stable across academic years
- No significant gender-based disparities in psychological outcomes

### 4. **Data Stability & Consistency**
- Burnout remains stable across academic years with minimal variation
- Strong and consistent patterns in mental health relationships
- Tight correlations suggest reliable, non-random relationships

### 5. **Recommendations for Modeling**
- **Problem Type**: Classification (predicting risk_level) with class imbalance handling
- **Feature Engineering**: Consider combining stress, anxiety, and burnout as a "mental health distress" composite feature
- **Model Considerations**: Use weighted loss functions or resampling techniques (SMOTE) to address class imbalance
- **Priority Features**: Focus on burnout, depression, stress, and mental health index as primary predictors

### 6. **Next Steps**
